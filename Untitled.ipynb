```python
import pandas as pd
import numpy as np
import os
import umap
reducer = umap.UMAP()
```

```python
labels = [
    'label:OR_standing',
    'label:SITTING',
    'label:LYING_DOWN',
    'label:FIX_running',
    'label:FIX_walking',
    'label:BICYCLING'
]
def select_columns(Y_train):
    y=[]
    for a in Y_train.values:
        y.append(labels[np.argmax(a)])
    return y
    

extrasensory_folder="android/extrasensory/"
extrasensory_folder1="android/client/app/src/main/assets/data/extrasensory1/"

def concatenateXY(folder):
    X_test=pd.read_csv(extrasensory_folder+folder+"/x_test.csv") 
    Y_test=pd.read_csv(extrasensory_folder+folder+"/y_test.csv", usecols=labels) 

    X_train=pd.read_csv(extrasensory_folder+folder+"/x_train.csv") 
    Y_train=pd.read_csv(extrasensory_folder+folder+"/y_train.csv", usecols=labels) 

    os.makedirs(extrasensory_folder1+folder, exist_ok=True)
    
    X_train["y"]=select_columns(Y_train)
    X_test["y"]=select_columns(Y_test)
    X_train.to_csv(extrasensory_folder1+folder+"/train.csv", index=False)
    X_test.to_csv(extrasensory_folder1+folder+"/test.csv", index=False)
    return [X_train,X_test]
    
    
 
from os import listdir

def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]
```
